### ЗАДАЧА: Плейлист плеера

Музыкальный плеер хранит треки в порядке воспроизведения.
Во время работы пользователь может:
- добавить трек в конец,
- вставить трек после указанного,
- удалить трек,
- переключиться на следующий или предыдущий трек.

НЕОБХОДИМО РЕАЛИЗОВАТЬ:

1. Класс `TrackNode` с полями:
   - `track_id`,
   - `title`,
   - `prev`,
   - `next`.

2. Класс `Playlist`, который хранит:
   - `head`,
   - `tail`,
   - `current` (текущий трек).

3. Методы:
   - `append_track(track_id, title)`,
   - `insert_after(target_id, track_id, title) -> bool`,
   - `remove_track(track_id) -> bool`,
   - `next_track() -> Optional[str]`,
   - `prev_track() -> Optional[str]`,
   - `current_track() -> Optional[str]`,
   - `to_list() -> list[str]`.

4. Выполнить команды из `actions` и вывести:
   - стартовый плейлист,
   - результаты операций,
   - финальный плейлист,
   - текущий трек в конце.


```python
from dataclasses import dataclass
from typing import Optional


initial_tracks = [
    ("T-1", "Wake Up"),
    ("T-2", "Ocean Drive"),
    ("T-3", "City Lights"),
    ("T-4", "After Rain"),
]

actions = [
    ("append", "T-5", "Night Ride"),
    ("insert_after", "T-2", "T-6", "Lo-Fi River"),
    ("next",),
    ("next",),
    ("remove", "T-3"),
    ("prev",),
    ("remove", "T-1"),
    ("insert_after", "NOPE", "T-7", "Ghost Song"),
    ("append", "T-8", "Morning Light"),
]


@dataclass
class TrackNode:
    track_id: str
    title: str
    prev: Optional["TrackNode"] = None
    next: Optional["TrackNode"] = None


class Playlist:
    def __init__(self):
        # TODO 1: создать head, tail, current
        # TODO 2: в пустом плейлисте они должны быть None
        self.head = None
        self.tail = None
        self.current = None

    def append_track(self, track_id: str, title: str) -> None:
        # TODO 1: создать новый узел
        # TODO 2: если плейлист пуст, сделать узел head/tail/current
        # TODO 3: иначе добавить узел после tail
        # TODO 4: обновить tail
        new = TrackNode(track_id, title)
        if self.head == None:
            self.head = new
            self.tail = new
            self.current = new
        else:
            self.tail.next = new
            new.prev = self.tail
            self.tail = new



    def insert_after(self, target_id: str, track_id: str, title: str) -> bool:
        # TODO 1: найти узел с target_id
        # TODO 2: если не найден, вернуть False
        # TODO 3: создать новый узел и вставить его после target
        # TODO 4: корректно обновить prev/next у соседей
        # TODO 5: если вставка в конец, обновить tail
        # TODO 6: вернуть True
        curr = self.head
        while curr:
            if curr.track_id == target_id:
                new = TrackNode(track_id=track_id, title=title)
                if curr == self.tail:
                    self.tail.next = new
                    new.prev = self.tail
                    self.tail = new
                else:
                    new.next = curr.next
                    new.prev = curr
                    curr.next = new
                    new.next.prev = new
                return True
            curr = curr.next
        return False

    def remove_track(self, track_id: str) -> bool:
        curr = self.head
        # TODO 1: найти узел с track_id
        # TODO 2: если не найден, вернуть False
        # TODO 3: перепривязать соседей (prev.next и next.prev)
        # TODO 4: если удаляется head или tail, обновить их
        # TODO 5: если удаляется current:
        #   - current = next, если next есть
        #   - иначе current = prev
        # TODO 6: если список стал пустым, current тоже должен стать None
        # TODO 7: вернуть True
        while curr:
            if curr.track_id == track_id:
                if curr == self.head:
                    self.head = curr.next
                    curr = self.head
                elif curr == self.tail:
                    self.tail = curr.prev
                    curr = self.tail
                elif curr == self.current:
                    a = curr.prev
                    a.next = curr.next
                    curr.next.prev = a
                    if self.current.next:
                        self.current = a.next
                    else:
                        self.current = a
                else:
                    a = curr.prev
                    a.next = curr.next
                    a.next.prev = a
                    
                    
                if self.head is None:
                    self.current = None
                return True
            curr = curr.next

    def next_track(self) -> Optional[str]:
        # TODO 1: если current отсутствует, вернуть None
        # TODO 2: если у current есть next, сдвинуть current на next
        # TODO 3: вернуть название текущего трека
        if self.current is None:
            return None
        if self.current.next:
            self.current = self.current.next
            return self.current.title

    def prev_track(self) -> Optional[str]:
        # TODO 1: если current отсутствует, вернуть None
        # TODO 2: если у current есть prev, сдвинуть current на prev
        # TODO 3: вернуть название текущего трека
        if self.current is None:
            return None
        if self.current.prev:
            self.current = self.current.prev
            return self.current.title

    def current_track(self) -> Optional[str]:
        # TODO: вернуть current.title, если current есть, иначе None
        if self.current is not None:
            return self.current.title
        return None

    def to_list(self) -> list[str]:
        # TODO 1: пройти от head до tail
        # TODO 2: собрать строки формата "track_id:title"
        # TODO 3: вернуть список
        curr = self.head
        arr = []
        while curr:
            arr.append(f'{curr.track_id}:{curr.title}')
            curr = curr.next
        return arr

playlist = Playlist()
for track_id, title in initial_tracks:
    playlist.append_track(track_id, title)

print("Стартовый плейлист:", playlist.to_list())
print("Сейчас играет:", playlist.current_track())

for action in actions:
    if action[0] == "append":
        _, track_id, title = action
        playlist.append_track(track_id, title)
        print("append ->", f"{track_id}:{title}")
    elif action[0] == "insert_after":
        _, target_id, track_id, title = action
        print("insert_after ->", target_id, playlist.insert_after(target_id, track_id, title))
    elif action[0] == "remove":
        _, track_id = action
        print("remove ->", track_id, playlist.remove_track(track_id))
    elif action[0] == "next":
        print("next ->", playlist.next_track())
    elif action[0] == "prev":
        print("prev ->", playlist.prev_track())


print("Финальный плейлист:", playlist.to_list())
print("Сейчас играет:", playlist.current_track())
```
