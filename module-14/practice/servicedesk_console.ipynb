# ЗАДАНИЕ: Консольное приложение для учета заявок на обслуживание оборудования

Необходимо разработать консольное приложение ServiceDesk Console для учета заявок на техническое обслуживание оборудования в организации.

В организации сотрудники используют различное оборудование: ноутбуки, компьютеры, принтеры, роутеры, сканеры и другую технику. Если оборудование выходит из строя или требует обслуживания, сотрудник оставляет заявку. Приложение должно позволять вести учет таких заявок, отслеживать их статус и получать краткую статистику.

## Цель работы:
Создать консольную программу, которая позволяет добавлять, просматривать, искать, редактировать и удалять заявки на обслуживание оборудования, а также сохранять данные между запусками программы.

## Функциональные требования:

### 1. Главное меню

При запуске программы пользователь должен видеть меню:

```txt
=== ServiceDesk Console ===

1. Добавить заявку
2. Показать все заявки
3. Найти заявку
4. Изменить статус заявки
5. Удалить заявку
6. Показать статистику
7. Сохранить данные
0. Выход
```
Пользователь выбирает действие путем ввода номера пункта меню.

### 2. Добавление заявки

Программа должна позволять добавить новую заявку.

При создании заявки пользователь вводит следующие данные:

```txt
- ФИО сотрудника
- отдел сотрудника
- тип оборудования
- инвентарный номер оборудования
- описание проблемы
- приоритет заявки
- статус заявки
```

**Приоритет заявки может принимать одно из значений:**

- низкий
- средний
- высокий

**Статус заявки может принимать одно из значений:**

- новая
- в работе
- выполнена

**Каждой заявке должен автоматически присваиваться уникальный номер.**

**Пример заявки:**
```txt
Номер заявки: 1
Сотрудник: Иванов Иван Иванович
Отдел: Бухгалтерия
Оборудование: Принтер
Инвентарный номер: PR-1023
Описание проблемы: Не печатает документы
Приоритет: высокий
Статус: новая
```

### 3. Просмотр всех заявок

Программа должна выводить список всех созданных заявок.

Для каждой заявки необходимо отображать:
```txt
- номер заявки
- ФИО сотрудника
- отдел
- тип оборудования
- инвентарный номер
- описание проблемы
- приоритет
- статус
```

Если заявок нет, программа должна вывести сообщение:

`Заявки отсутствуют.`

### 4. Поиск заявок

Программа должна позволять искать заявки по разным параметрам.

Необходимо реализовать поиск:
```txt
- по ФИО сотрудника
- по отделу
- по статусу
- по инвентарному номеру оборудования
```

При выборе пункта `Найти заявку` пользователь должен выбрать критерий поиска, затем ввести значение для поиска.

Пример меню поиска:
```txt
1. Поиск по сотруднику
2. Поиск по отделу
3. Поиск по статусу
4. Поиск по инвентарному номеру
```
Если подходящие заявки найдены, программа выводит их список.
Если заявок не найдено, программа выводит сообщение:

`Заявки не найдены.`

### 5. Изменение статуса заявки

Программа должна позволять изменить статус существующей заявки.

Для изменения статуса пользователь вводит номер заявки, после чего выбирает новый статус:
```txt
1. новая
2. в работе
3. выполнена
```

Если заявка с указанным номером существует, программа изменяет ее статус и выводит сообщение об успешном изменении.

Если заявка не найдена, программа выводит сообщение:

`Заявка с указанным номером не найдена.`

### 6. Удаление заявки

Программа должна позволять удалить заявку.

Для удаления пользователь вводит номер заявки.

Если заявка с таким номером существует, программа удаляет ее и выводит сообщение:

`Заявка успешно удалена.`

Если заявка не найдена, программа выводит сообщение:

`Заявка с указанным номером не найдена.`

### 7. Статистика

Программа должна выводить краткую статистику по заявкам:

- общее количество заявок
- количество новых заявок
- количество заявок в работе
- количество выполненных заявок
- количество заявок с высоким приоритетом

Пример вывода:
```txt
=== Статистика ===

Всего заявок: 10
Новые: 3
В работе: 4
Выполнены: 3
С высоким приоритетом: 2
```

### 8. Сохранение данных

Программа должна сохранять данные о заявках в файл.

Формат файла можно выбрать самостоятельно:

- txt
- csv
- json

После сохранения программа должна вывести сообщение:

`Данные успешно сохранены.`

### 9. Загрузка данных

При запуске программы необходимо загружать ранее сохраненные заявки из файла, если такой файл существует.

Если файл отсутствует, программа начинает работу с пустым списком заявок.

### 10. Выход из программы

При выборе пункта `0. Выход` программа должна завершить работу.

Перед выходом желательно выполнить автоматическое сохранение данных в файл.

Требования к обработке ошибок:

1. Программа не должна завершаться аварийно при некорректном вводе пользователя.
2. Если пользователь вводит неправильный пункт меню, необходимо вывести сообщение:

`Некорректный выбор. Попробуйте снова.`

3. Если пользователь вводит несуществующий номер заявки, необходимо вывести сообщение:

`Заявка с указанным номером не найдена.`

4. Если пользователь вводит недопустимый статус или приоритет, программа должна попросить повторить ввод.

Требования к структуре программы:

В программе рекомендуется использовать:

- класс или структуру для хранения данных заявки
- список или коллекцию для хранения всех заявок
- отдельные функции или методы для:
  - вывода меню
  - добавления заявки
  - просмотра заявок
  - поиска заявок
  - изменения статуса
  - удаления заявки
  - вывода статистики
  - сохранения данных
  - загрузки данных

Пример структуры заявки:

Заявка:
- id
- employeeName
- department
- equipmentType
- inventoryNumber
- problemDescription
- priority
- status

Дополнительные требования:

1. Номер заявки должен быть уникальным.
2. После удаления заявки номера остальных заявок не должны изменяться.
3. При добавлении новой заявки номер должен быть больше, чем у всех уже существующих заявок.
4. Поиск должен работать без учета регистра букв.
5. Данные должны сохраняться между запусками программы.

Результат выполнения работы:

В результате должно быть создано консольное приложение ServiceDesk Console, которое позволяет вести учет заявок на обслуживание оборудования в организации и сохранять информацию о заявках между запусками программы.

````python
from dataclasses import dataclass


@dataclass
class Case:
    id: str
    employeeName: str
    department: str
    equipmentType: str
    inventoryNumber: int
    problemDescription: str
    priority: str
    status: str = 'новая'
    def set_status(self, status):
        if status not in {'новая', 'в работе', 'выполнена'}:
            print('некорректный статус!')
            return
        self.status = status


class Saver:
    @staticmethod
    def save(l: list[Case]):
        with open('report.txt', 'w', encoding='utf-8') as f:
            f.write('       == Формат заявок: "id | employeeName | department | equipmentType | inventoryNumber | problemDescription | priority | status" ==\n')
            for case in l:
                f.write(f'{case.id} | {case.employeeName} | {case.department} | {case.equipmentType} | {case.inventoryNumber} | {case.problemDescription} | {case.priority} | {case.status}\n')
        print('Данные успешно сохранены')
    def pull(service):
        res = []
        with open('report.txt', 'r', encoding='utf-8') as f:
            lines = f.read().split('\n')[1:-1]
            for line in lines:
                args = line.split(' | ')
                case = Case(args[0], args[1], args[2], args[3], int(args[4]), args[5], args[6], args[7])
                res.append(case)
        service._pull(res)

class CaseService:
    def __init__(self):
        self.cases = {}
        self.max_number = 1
    def add_case(self, employeeName, department, equipmentType, inventoryNumber, problemDescription, priority):
        if priority not in {"низкий", "высокий", "средний"}:
            print('некорректный приоритет!')
            return
        case_id = "CASE-" +  str(self.max_number)
        self.cases[case_id] = Case(case_id, employeeName, department, equipmentType, inventoryNumber, problemDescription, 
                                   priority)
        self.max_number += 1
    
    def show_cases(self):
        if not self.cases:
            print('Заявки отсутствуют.')
            return
        for v in self.cases.values():
            print(f'номер заявки: {v.id}')
            print(f'ФИО сотрудника: {v.employeeName}')
            print(f'отдел: {v.department}')
            print(f'тип оборудования: {v.equipmentType}')
            print(f'инвентарный номер: {v.inventoryNumber}')
            print(f'описание проблемы: {v.problemDescription}')
            print(f'приоритет: {v.priority}')
            print(f'статус: {v.status}')
        
    def find_by_fio(self, target):
        print(f'Заявки с работником {target}:')
        is_empty = True
        for v in self.cases.values():
            if v.employeeName == target:
                print(v)
                is_empty = False
        if is_empty:
            print('Заявки не найдены')
    
    def find_by_department(self, target):
        print(f'Заявки с отделом {target}:')
        is_empty = True
        for v in self.cases.values():
            if v.department == target:
                print(v)
                is_empty = False
        if is_empty:
            print('Заявки не найдены')
        
    def find_by_status(self, target):
        print(f'Заявки со статусом {target}:')
        is_empty = True
        for v in self.cases.values():
            if v.status == target:
                print(v)
                is_empty = False
        if is_empty:
            print('Заявки не найдены')
    
    def find_by_number(self, target):
        print(f'Сотрудники с инвентарным номером {target}:')
        is_empty = True
        for v in self.cases.values():
            if v.inventoryNumber == target:
                print(v)
                is_empty = False
        if is_empty:
            print('Заявки не найдены')
    
    def change_status(self, case_id, status):
        case: Case = self.cases.get(case_id)
        if case == None:
            print('Заявка с указанным номером не найдена')
            return
        case.set_status(status)
    
    def delete_case(self, case_id):
        if case_id not in self.cases:
            print('Заявка с указанным номером не найдена')

        del self.cases[case_id]
    
    def show_statistic(self):
        print('         == Статистика ==')
        print(f'Общее количество заявок: {len(list(self.cases.values()))}')
        new = 0
        in_work = 0
        done = 0
        high_priority = 0
        for case in self.cases.values():
            if case.status == 'новая':
                new += 1
            if case.status == 'в работе':
                in_work += 1
            if case.status == 'выполнена':
                done += 1
            if case.priority == 'высокий':
                high_priority += 1
        print(f'количество новых заявок: {new}')
        print(f'количество заявок в работе: {in_work}')
        print(f'количество выполненных заявок: {done}')
        print(f'количество заявок с высоким приоритетом: {high_priority}')
    
    def _pull(self, arr):
        self.cases.clear()
        for case in arr:
            self.cases[case.id] = case
            


class Menu:
    def __init__(self):
        self.service = CaseService()
    def create_case(self):
        employeeName = input('введите имя сотрудника: ')
        department = input('Введите департамент: ')
        equipmentType = input('Введите тип оборудования: ')
        inventoryNumber = input('Введите инвентарный номер: ')
        problemDescription = input('Введите описание проблемы: ')
        priority = input('Введите приоритет: ')
        self.service.add_case(employeeName, department, equipmentType, inventoryNumber, problemDescription, priority)
    
    def find(self):
        ch = input('Выберите действие: ')
        if ch == '1':
            target = input('Введите имя: ')
            self.service.find_by_fio(target)
        if ch == '2':
            target = input('Введите отдел: ')
            self.service.find_by_department(target)
        if ch == '3':
            target = input('Введите статус: ')
            self.service.find_by_status(target)
        if ch == '4':
            target == int(input('Введите инвентарный номер: '))
            self.service.find_by_number(target)
    
    def change_status(self):
        case_id = input('Введите id заявки: ')
        if case_id not in self.service.cases:
            print('Заявка с указанным номером не найдена.')
            return
        st = input('Введите новый статус: ')
        self.service.change_status(case_id, st)

    def delete_status(self):
        case_id = input('Введите id заявки: ')
        if case_id not in self.service.cases:
            print('Заявка с указанным номером не найдена.')
            return
        self.service.delete_case(case_id)
    def main(self):
        while True:
            print('''== ServiceDesk Console ==
    1. добавить заявку
    2.показать все заявки
    3.найти заявку
    4.изменить статус заявки
    5.удалить заявку
    6.показать статистику
    7.сохранить данные
    8.Загрузить данные
    0.выход''')
            ch = input('Выберите действие: ')
            if ch == '1':
                self.create_case()
            elif ch == '2':
                self.service.show_cases()
            elif ch == '3':
                print("""1.Поиск по сотруднику
2.Поиск по отделу
3.Поиск по статусу
4.Поиск инвентарному номеру""")
                self.find()
            elif ch == '4':
                self.change_status()
            elif ch == '5':
                self.delete_status()
            elif ch == '6':
                self.service.show_statistic()
            elif ch == '7':
                Saver.save(list(self.service.cases.values()))
            elif ch == '8':
                Saver.pull(self.service)
            elif ch == '0':
                Saver.save(list(self.service.cases.values()))
                break
            else:
                print('Некорректный выбор. Попробуйте снова')

menu = Menu()
menu.main()
                


        


    








        

        




    
````
